```python
import pandas as pd
import numpy as np
import math
```

```python
tax_df = pd.read_excel("taxonomy.xlsx")
tax_df
```

```python
# Create taxonhierarchy
rank_list = list(tax_df.columns.values)
taxonhierarcy = pd.DataFrame(rank_list, columns=["rank_name"])
taxonhierarcy.insert(0, "rank_id", range(len(taxonhierarcy)))
taxonhierarcy.insert(2, "parent_id", (taxonhierarcy["rank_id"] - 1))
taxonhierarcy.loc[taxonhierarcy['parent_id'] == -1, 'parent_id'] = "NaN"
taxonhierarcy
```

```python
tax_list = tax_df.values.flatten().tolist()
tax_id_list = []
parent_tax_id_list = []
rank_id_list = []
taxon_name = []
parent_pos = 0
pos = 0
for taxStr in tax_list:
    if str(taxStr) != 'nan':
        if taxStr not in taxon_name:
            taxon_name.append(taxStr)
            parent_tax_id_list.append(parent_pos)
            tax_id_list.append(len(taxon_name))
            rank_id_list.append(pos)
            if pos == 12:
                parent_pos = 0
            else:
                parent_pos = len(taxon_name)
        else:
            parent_pos = taxon_name.index(taxStr) + 1
        pos = (pos + 1) % 13
    else:
        pos = (pos + 1) % 13

print(parent_tax_id_list)
print(rank_id_list)
print(taxon_name)
```

```python
taxonomy = pd.DataFrame({'taxonomy_id':tax_id_list, 'parent_taxon_id':parent_tax_id_list, 'rank_id':rank_id_list, 'taxon_name':taxon_name})
taxonomy
```

```python
datatoexcelTaxonomy = pd.ExcelWriter('taxonomyDummy.xlsx')

taxonomy.to_excel(datatoexcelTaxonomy, index=False)

datatoexcelTaxonomy.save()
```
