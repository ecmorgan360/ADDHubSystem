```python
collection_df = pd.read_excel('collection.xlsx')

# Separate out all collectors
collectors = collection_df['collectedBy'].unique().tolist()
separateCollectors = []
for collector in collectors:
    separateCollectors.append(collector.split(" and "))
separateCollectors = [item for sublist in separateCollectors for item in sublist]
finalCollectors = []
for collector in separateCollectors:
    finalCollectors.append(collector.split("; "))
finalCollectors = [item for sublist in finalCollectors for item in sublist]
finalCollectors = list(dict.fromkeys(finalCollectors))
finalCollectors
```

### person, role and person_role tables

```python
# Person_role
assigners = pd.concat([amr_df['assignedBy'], mnp_df['assignedBy']]).unique().tolist()
assignerDict = {'full_name':assigners, 'role':'Assigner'}
bioassays = bioassay_df['researcher'].unique().tolist()
bioassayDict = {'full_name':bioassays, 'role':'Bioassay Scientific Analyst'}
extractors = extract_df['submittedBy'].unique().tolist()
extractDict = {'full_name':extractors, 'role':'Extractor'}
fractionators = pd.concat([pure_df['submittedBy'], mnp_df['submittedBy']]).unique().tolist()
fractionDict = {'full_name':fractionators, 'role':'Fractionator'}
identifiers = collection_df['identifiedBy'].unique().tolist()
identifyDict = {'full_name':identifiers, 'role':'Identifier'}
collectorDict = {'full_name':finalCollectors, 'role':'Collector'}
person_role_df = pd.concat([pd.DataFrame(assignerDict), pd.DataFrame(bioassayDict), pd.DataFrame(extractDict), pd.DataFrame(fractionDict), pd.DataFrame(identifyDict), pd.DataFrame(collectorDict)], ignore_index=True)
print(assigners)
print(bioassays)
print(extractors)
print(fractionators)
print(person_role_df)
```

```python
# Person table
collector_df = pd.DataFrame(collectorDict)
people = pd.concat([amr_df['assignedBy'], amr_df['submittedBy'], mnp_df['assignedBy'], mnp_df['submittedBy'], collection_df['identifiedBy'], collector_df['full_name']]).unique().tolist()
person_df = pd.DataFrame(people, columns=['full_name'])
person_df.insert(0, 'person_id', range(1, 1 + len(person_df)))
person_df
```

```python
# Role table
roles = person_role_df['role'].unique().tolist()
role_df = pd.DataFrame(roles, columns=['role'])
role_df.insert(0, 'id', range(1, 1 + len(role_df)))
role_df
```

```python
# Person role table
person_role_df.insert(2, 'person_id', 1)
person_role_df.insert(3, 'id', 1)
roleIDs = role_df['id'].tolist()
roles = role_df['role'].tolist()
personIDs = person_df['person_id'].tolist()
names = person_df['full_name'].tolist()
for rID, r in zip(roleIDs, roles):
    person_role_df.loc[person_role_df.role == r, 'id'] = rID

for pID, p in zip(personIDs, names):
    person_role_df.loc[person_role_df.full_name == p, 'person_id'] = pID

person_role_df.drop(['full_name', 'role'], axis=1, inplace=True)
person_role_df.rename(columns={'person_id':'user_id', 'id':'role_id'}, inplace=True)

person_role_df
```

```python
# Person table:
# person_id, full_name, email_address, database_user, username, password
# Making email addresses based on person name
nameSeries = person_df['full_name']
emailList = []
initialList = []
passwordList = []
for index, value in nameSeries.items():
    email = ""
    initial = ""
    password = ""
    allNames = value.split(" ")
    for name in allNames:
        password = password + name[0:2].lower()
        initial = initial + name.lower()
        email = email + name[0].lower()
    email = email + "@ru.ac.za"
    password = password + "12345$"
    emailList.append(email)
    initialList.append(initial)
    passwordList.append(password)

emailSeries = pd.Series(emailList)
person_df.insert(2, 'email_address', emailSeries)
userSeries = pd.Series(initialList)
passwordSeries = pd.Series(passwordList)

# Saying only those who are assigners have current access to database
# Only those will have usernames and passwords
person_df.insert(3, 'database_user', "N")
person_df.loc[person_role_df['role_id'] == 1, 'database_user'] = "Y"
person_df.insert(4, 'username', userSeries)
person_df.loc[person_df['database_user'] == "N", 'username'] = None
person_df.insert(5, 'password', passwordSeries)
person_df.loc[person_df['database_user'] == "N", 'password'] = None
person_df
```

```python
# Need from extract_df, fraction and pure_df
# processID (AMRID or MNPID), supplierID, assignedBy, assignedDate, destination, sample_level (extract, fraction, pure comp)
processExtract = extract_df[["amrID", "supplierID", "assignedBy", "assignedDate", "destinationOfSample"]]
processExtract.insert(5, 'sample_level', "Extract")
processExtract.rename(columns={'amrID':'process_id'}, inplace=True)
processFraction = mnp_df[["mnpID", "supplierID", "assignedBy", "assignedDate", "destinationOfSample"]]
processFraction.insert(5, 'sample_level', "Fraction")
processFraction.rename(columns={'mnpID':'process_id'}, inplace=True)
processPure = pure_df[["amrID", "supplierID", "assignedBy", "assignedDate", "destinationOfSample"]]
processPure.insert(5, 'sample_level', "PureCompound")
processPure.rename(columns={'amrID':'process_id'}, inplace=True)
process = pd.concat([processExtract, processFraction, processPure], ignore_index=True)
process.rename(columns={'supplierID':'supplier_id', 'assignedDate':'date_assigned', 'destinationOfSample':'destination_of_sample'}, inplace=True)
process.insert(2, 'assigner_id', 1)
personIDs = person_df['person_id'].tolist()
names = person_df['full_name'].tolist()

for pID, p in zip(personIDs, names):
    process.loc[process.assignedBy == p, 'assigner_id'] = pID

process.drop(['assignedBy'], axis=1, inplace=True)
# Date must be converted to datetime format for the database
process['date_assigned'] = pd.to_datetime(process['date_assigned'])
process
```

### Mass spec raw table

```python
# Find where destination of sample is MS, but not LC/MS or LC-MS
ms_records = process[process.destination_of_sample.str.contains('MS', regex=False)]
ms_records = ms_records.drop(ms_records[ms_records.destination_of_sample.str.contains('LC/MS', regex=False)].index)
ms_records = ms_records.drop(ms_records[ms_records.destination_of_sample.str.contains('LC-MS', regex=False)].index)
ms_records.drop(['supplier_id', 'assigner_id', 'destination_of_sample', 'sample_level'], axis=1, inplace=True)

# Now make mass spec records for each of these
# massspec_id will be autocreated (incrementing by 1), process_id, date_requested will be same as date_assigned, cancelled all false, date_submitted will all be None, link_processed_spec all None
ms_records.insert(2, 'cancelled', False)
# ms_records.insert(3, 'link_processed_spec', None)
# ms_records.insert(4, 'date_submitted', None)
ms_records.rename(columns={'date_assigned':'date_requested'}, inplace=True)
# Date must be converted to datetime format for the database
ms_records['date_requested'] = pd.to_datetime(ms_records['date_requested'])
ms_records
```

```python
# Get 80% of these records and make mass spec info as if they were already uploaded
massProcessed = ms_records.sample(frac=0.8).reset_index(drop=True)

#Then create fake link
fake = Faker()
Faker.seed(1234)
random.seed(1234)

class Provider(BaseProvider):
    def link_processed_spec(self):
        return "/ms/" + fake.word()

# Add the Provider to our faker object
fake.add_provider(Provider)

# Go up by one for AMR ID
# Supplier ID - based on date of collection, so must be same or earlier year than submittedDate
#               some TIC, some SAF, Some strains (e.g. S 20 20), MIAP_155, HMF-1, MBA_F406_PDM
# Date assigned - do a week after submittedDate
# Source ID - TIC number, MIAP etc - based on Supplier ID (start with this, then make Supplier ID)
# Source type - based on source ID (microbial vs sponge vs stromatolite)
# Permit - RES and year before or same as source ID

def create_ms_data(fake, no_of_rows):
    columns = ['link_processed_spec']
    data =  {column: [getattr(fake, column)() for _ in range(no_of_rows)] for column in columns}
    df = pd.DataFrame(data=data)
    df = df[columns]

    return df

original_df = create_ms_data(fake, len(massProcessed))
totalms = pd.concat([massProcessed, original_df], axis=1)
# Make date submitted 2 weeks after date requested
totalms.insert(4, 'date_submitted', totalms['date_requested'])
# Date submitted - do 3 weeks after previous assignedDate
totalms['date_submitted'] = totalms['date_submitted'] + timedelta(days=14)

newms = ms_records.merge(totalms, on=['process_id', 'date_requested', 'cancelled'], how='left')
newms
```
